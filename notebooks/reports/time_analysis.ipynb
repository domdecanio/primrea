```python
# Time steps:
date_08 = pd.Timestamp('2008-01-01')
date_09 = pd.Timestamp('2009-01-01')
date_10 = pd.Timestamp('2010-01-01')
date_11 = pd.Timestamp('2011-01-01')
date_12 = pd.Timestamp('2012-01-01')
date_13 = pd.Timestamp('2013-01-01')
date_14 = pd.Timestamp('2014-01-01')
date_15 = pd.Timestamp('2015-01-01')
date_16 = pd.Timestamp('2016-01-01')
date_17 = pd.Timestamp('2017-01-01')
date_18 = pd.Timestamp('2018-01-01')
date_19 = pd.Timestamp('2019-01-01')
date_20 = pd.Timestamp('2020-01-01')
date_21 = pd.Timestamp('2021-01-01')
date_22 = pd.Timestamp('2022-01-01')
date_23 = pd.Timestamp('2023-01-01')
```

```python
def count_pubs(tethys_df):
    '''
    '''
    tech_lst_1 = list(tethys_df['technologyType'])
    tech_indicator_1 = list()
    tech_indicator_2 = list()
    
    for i in tech_lst_1:
        if len(i) == 0:
            tech_indicator_1.append(0)
            tech_indicator_2.append(0)
        else:
            tech_indicator_1.append(1)
    
            if i[0] == 'Wave': 
                tech_indicator_2.append(1)
            elif i[0] == 'Current': 
                tech_indicator_2.append(2)
            else:
                tech_indicator_2.append(3)
     
            
    tethys_df['has_technologyType']=tech_indicator_1
    tethys_df['typ_technologyType']=tech_indicator_2

    df = pd.DataFrame(tethys_df['typ_technologyType'].value_counts())
    
    return df
```

```python
periods = [
    date_08,
    date_09,
    date_10,
    date_11,
    date_12,
    date_13,
    date_14,
    date_15,
    date_16,
    date_17,
    date_18,
    date_19,
    date_20,
    date_21,
    date_22,
    date_23 ]

```

```python
# Tethys

t_results_0 = list()
t_results_1 = list()
t_results_2 = list()
t_results_3 = list()

for i in range(1, len(periods)):
    df = count_pubs(tethys_df[(tethys_df_2['originationDate']>=periods[i-1]) & (tethys_df_2['originationDate']<periods[i])])
    t_results_0.append(df['typ_technologyType'][0])
    t_results_1.append(df['typ_technologyType'][1])
    t_results_2.append(df['typ_technologyType'][2])
    #t_results_3.append(df['typ_technologyType'][3])
```
